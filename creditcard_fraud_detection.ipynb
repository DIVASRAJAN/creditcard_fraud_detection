```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/content/creditcard.csv')
df
```

```python
df.info()
```

```python
df.isna().sum()
```

```python
df.drop(['Time'],axis=1,inplace=True)
```

```python
df
```

```python
df.drop_duplicates(inplace=True)
```

```python
df.shape
```

```python
df['Class'].value_counts()
```

```python
x= df.iloc[:,:-1]
y=df.iloc[:,-1]
```

```python
from imblearn.over_sampling import SMOTE

smote = SMOTE(random_state=42)
x_re, y_re = smote.fit_resample(x, y)
```

```python
y_re.value_counts()
```

```python
sns.heatmap(df.corr())
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler=MinMaxScaler()
scaler.fit(x_re)
x_re=scaler.transform(x_re)
```

### Feature Selection

```python
from sklearn.feature_selection import SelectKBest,chi2

chi=SelectKBest(chi2,k=25)
best=chi.fit_transform(x_re,y_re)
print(best.shape)
x_indes=chi.get_support(indices=True)
print(df.columns[x_indes])
```

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test=train_test_split(best,y_re,test_size=0.30,random_state=42)
```

### Random Forest Classifier

```python
from sklearn.metrics import accuracy_score, classification_report
from sklearn.ensemble import RandomForestClassifier


model = RandomForestClassifier(n_estimators=100, criterion='gini', random_state=42)

model.fit(x_train,y_train)
y_pred = model.predict(x_test)

accuracy = accuracy_score(y_test, y_pred)
report = classification_report(y_test, y_pred)

print(accuracy)
print(report)
```

### Model Saving

```python
import pickle

with open('credit_card.pickle', 'wb') as dump_var:
    pickle.dump(model, dump_var)
```

### Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report

logistic= LogisticRegression()
logistic.fit(x_train, y_train)
y_pred = logistic.predict(x_test)

accuracy1 = accuracy_score(y_test, y_pred)
report1 = classification_report(y_test, y_pred)

print(accuracy1)
print(report1)
```

### Naive Bayes

```python
from sklearn.naive_bayes import GaussianNB

gauss = GaussianNB()
gauss.fit(x_train, y_train)
y_pred = gauss.predict(x_test)

accuracy2 = accuracy_score(y_test, y_pred)
report2 = classification_report(y_test, y_pred)

print(accuracy2)
print(report2)
```

### Gradient Boosting

```python
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.metrics import accuracy_score, classification_report

gradient = GradientBoostingClassifier(random_state=10)

gradient.fit(x_train, y_train)
y_pred = gradient.predict(x_test)

accuracy3 = accuracy_score(y_test, y_pred)
report3 = classification_report(y_test, y_pred)

print(accuracy3)
print(report3)
```

### Adaboost Classifier

```python
from sklearn.ensemble import AdaBoostClassifier


adaboost = AdaBoostClassifier(random_state=42)
adaboost.fit(x_train, y_train)

y_pred = adaboost.predict(x_test)

accuracy4 = accuracy_score(y_test, y_pred)
report4 = classification_report(y_test, y_pred)

print(accuracy4)
print(report4)
```

```python
name=['random','logistic','naive','gradient','adaboost']
result=[accuracy,accuracy1,accuracy2,accuracy3,accuracy4]
```

```python
bars = plt.bar(name, result)
for bar, value in zip(bars, result):
    plt.text(bar.get_x() + bar.get_width() / 2, bar.get_height(), round(value,2),
             ha='center', va='bottom')

plt.show()
```
